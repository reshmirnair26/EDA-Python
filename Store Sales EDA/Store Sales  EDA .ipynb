#### Rename columns

```python
data.rename(columns={'Row ID':'Row_ID','Order ID':'Order_ID','Order Date':'Order_Date','Ship Date':'Ship_Date','Ship Mode':'Ship_Mode','Customer ID':'Customer_ID','Customer Name':'Customer_Name','Postal Code':'Postal_Code','Product ID':'Product_ID','Product Name':'Product_Name'},inplace=True)
data.sample()
```

```python
data.describe().T
```

#### Null value

```python
data.isnull().mean()*100
```

#### Dropping Null value

```python
data.dropna(inplace=True)
```

```python
data.isnull().mean()*100
```

#### Duplicate value

```python
data.duplicated().sum()
```

#### Delay Date

```python
data['Delay']=data['Order_Date']-data['Ship_Date']
data.Delay.value_counts
```

```python
# Delay dtype covert to string
```

```python
data['Delay'] = data['Delay'].astype('str')
data.sample(5)
```

```python
# Delay elements replaced by digits only
```

```python
data['Delay1']=data.Delay.str.replace("\D","",regex=True)
data.head()
```

```python
#  delay1 dtype convert to integer
```

```python
data['Delay1'] = data['Delay1'].astype('int')
data.info()
```

#### Day of week

```python
data['WeekDay'] = data.Order_Date.dt.day_of_week
```

```python
data.head()
```

```python
data.WeekDay.replace({0:'Sunday', 1:'Monday', 2:'Tuesday',3:'wednesday',4:'Thursday',5:'Friday',6:'Saturday'}, inplace=True)
data.head()
```

```python
data.drop(columns='Delay',inplace=True)
```
