### For loop to plot subplot

```python
for i in data.columns:
    if data[i].dtype!='object':
        sns.histplot(x=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtype !='object':
        sns.kdeplot(data=data[i])
        plt.show()
```

### Subplot in matplotlib

```python
plt.figure(figsize=(6,10))
plt.subplot(2,2,1)
sns.histplot(data=data,x='Survived')

plt.subplot(2,2,2)
sns.histplot(data=data,x='Sex')

plt.subplot(2,2,3)
sns.histplot(data=data,x='Age')

plt.subplot(2,2,4)
sns.histplot(data=data,x='Pclass')

plt.show()
```

### Pie chart

```python
SibSp_Fare=data.groupby(['SibSp'])['Fare'].sum()
SibSp_Fare.plot.pie(legend=True,autopct='%1.2f%%');
```

```python
Embarked_Fare=data.groupby(['Embarked'])['Fare'].sum()
Embarked_Fare.plot.pie(legend=True,autopct='%1.2f%%');
```

```python
Embarked_fare_mean=data.groupby(['Embarked'])['Fare'].mean()
Embarked_fare_mean.plot.bar();
```
